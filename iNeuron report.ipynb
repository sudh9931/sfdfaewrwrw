```python
import pandas as pd
pd.set_option('display.max_rows', 500) 
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 500)
```

```python
#1.Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave
df_perf = pd.read_excel('AgentPerformance.xlsx',skiprows=1,engine="openpyxl")
df_perf['Week']=pd.to_datetime(df_perf.Date).dt.isocalendar().week
df_perf['Average Rating']=df_perf['Average Rating'].astype(float)
df_perf['Total Chats']=df_perf['Total Chats'].astype(int)
df_perf[(df_perf['Total Chats'] > 0) ][['Agent Name','Average Rating','Date','Week']].groupby \
   (['Agent Name','Week'])['Average Rating'].mean()



```

```python
#2.Total working days for each agents
df_perf = pd.read_excel('AgentPerformance.xlsx',skiprows=1,engine="openpyxl")
df_perf['Total Chats']=df_perf['Total Chats'].astype(int)
df_perf[(df_perf['Total Chats'] > 0)][['Agent Name']].value_counts()


```

```python
#3. Total query that you hvae taken
df_perf = pd.read_excel('/Users/nikhilgopalakrishnan/Downloads/AgentPerformance.xlsx',skiprows=1,engine="openpyxl")
df_perf['Total Chats']=df_perf['Total Chats'].astype(int)
df_perf[(df_perf['Total Chats'] > 0)].groupby('Agent Name')['Total Chats'].agg(sum)
```
